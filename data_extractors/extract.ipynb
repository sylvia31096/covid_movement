```python
import pandas as pd
```

```python
cc_df = pd.read_csv('2019-population_census-report-per-constituency.csv')
```

```python
cc_df = cc_df[['sub_county', 'male_population',
       'female_population', 'total_population', 'population_density']]
cc_df.columns = ['region_name','male_population','female_population',
              'total_population','population_density']
for i in cc_df.columns:
    cc_df[i] =cc_df[i].astype(str).str.replace(',','')
cc_df.head()
```

```python
c_df = pd.read_csv('2019-population_census-report-per-county.csv')
```

```python
c_df = c_df[['county','male_population','female_population',
    'intersex_population','total_population','population_density']]
```

```python
c_df.columns=['region_name','male_population','female_population',
    'intersex_population','total_population','population_density']
for i in c_df.columns:
    c_df[i] =c_df[i].astype(str).str.replace(',','')
c_df.head()
```

```python
df = c_df.append(cc_df)
df.tail()
```

```python
import numpy as np
```

```python
df = df.replace('..',np.nan)
df = df.replace('-',np.nan)

df = df.where(pd.notnull(df), None)

df.tail()
```

```python
df.to_csv('region_pop.csv',index=False)
```
